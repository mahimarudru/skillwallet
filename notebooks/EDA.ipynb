```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
data = pd.read_csv("../data/patient_data.csv")
data.head()
```

## Dataset Overview

This section provides information about the dataset structure,
including column names, data types, and missing values.


```python
data.info()
```

```python
data.describe(include="all")
```

## Missing Values Analysis

```python
data.isnull().sum()
```

## Distribution of Hypertension Stages

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Stages", data=data)
plt.xticks(rotation=45)
plt.title("Distribution of Hypertension Stages")
plt.show()
```

## Age Group Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Age", data=data)
plt.title("Age Distribution")
plt.show()
```

## Relationship Between Systolic BP and Hypertension Stage

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Systolic", hue="Stages", data=data)
plt.xticks(rotation=45)
plt.title("Systolic Blood Pressure vs Stage")
plt.show()
```

## Relationship Between Diastolic BP and Hypertension Stage

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Diastolic", hue="Stages", data=data)
plt.xticks(rotation=45)
plt.title("Diastolic Blood Pressure vs Stage")
plt.show()
```

## Feature Correlation Analysis

```python
encoded = pd.get_dummies(data)

plt.figure(figsize=(12,8))
sns.heatmap(encoded.corr(), cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Observations

- Hypertension stages strongly correlate with systolic and diastolic blood pressure ranges.
- Certain age groups appear more frequently in higher hypertension stages.
- Blood pressure features show strong relationships with the target variable.
- The dataset appears structured according to medical classification rules.
